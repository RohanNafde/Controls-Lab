```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sensor = []
time = []

data_file1 = 'E:\EE core\Sem 5\Controls Lab\Expt1\sensor.txt'
data_file2 = 'E:\EE core\Sem 5\Controls Lab\Expt1\\time.txt'

with open(data_file1, 'r') as file:
    for line in file:
        data = line.strip().split()
        sensor.append(float(data[0]))

with open(data_file2, 'r') as file:
    for line in file:
        data1 = line.strip().split()
        time.append(float(data1[0]))

print(time)
```

```python
plt.plot(time, sensor)
plt.ylabel("Sensor Value")
plt.xlabel("Time (ms)")
plt.show()
```
